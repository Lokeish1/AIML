```python
# Basic Python Program - Factorial of a number
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example Usage
number = 5
print("Factorial of", number, "is:", factorial(number))
```

```python
from collections import deque

# Sample Graph represented as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Breadth-First Search (BFS)
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            queue.extend(neighbor for neighbor in graph[node] if neighbor not in visited)
    print()

# Depth-First Search (DFS)
def dfs(graph, start, visited=set()):
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Iterative Deepening DFS (IDFS)
def idfs(graph, start, max_depth):
    def dls(node, depth):
        if depth == 0:
            return [node]
        elif depth > 0:
            path = []
            for neighbor in graph.get(node, []):
                path.extend(dls(neighbor, depth - 1))
            return path
        return []

    for depth in range(max_depth):
        visited = set(dls(start, depth))
        for node in visited:
            print(node, end=" ")
        print()

# Example Usage
print("BFS Traversal:")
bfs(graph, 'A')

print("DFS Traversal:")
dfs(graph, 'A')
print()

print("IDFS Traversal up to Depth 2:")
idfs(graph, 'A', max_depth=2)
```

```python
from queue import PriorityQueue

graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 3), ('E', 1)],
    'C': [('F', 5)],
    'D': [],
    'E': [('F', 2)],
    'F': []
}

heuristics = {'A': 6, 'B': 4, 'C': 5, 'D': 6, 'E': 2, 'F': 0}

def a_star(graph, heuristics, start, goal):
    queue = PriorityQueue()
    queue.put((0, start))
    costs = {start: 0}
    path = {}

    while not queue.empty():
        _, current = queue.get()
        if current == goal:
            break
        for neighbor, cost in graph[current]:
            new_cost = costs[current] + cost
            if neighbor not in costs or new_cost < costs[neighbor]:
                costs[neighbor] = new_cost
                priority = new_cost + heuristics[neighbor]
                queue.put((priority, neighbor))
                path[neighbor] = current

    current = goal
    route = []
    while current in path:
        route.append(current)
        current = path[current]
    route.append(start)
    route.reverse()
    return route

print("A* Path from A to F:", a_star(graph, heuristics, 'A', 'F'))
```

```python
countries = ['A', 'B', 'C', 'D']
neighbors = {'A': ['B', 'C'], 'B': ['A', 'D'], 'C': ['A', 'D'], 'D': ['B', 'C']}
colors = ['Red', 'Green', 'Blue']

def is_safe(country, color, assignment):
    for neighbor in neighbors[country]:
        if assignment.get(neighbor) == color:
            return False
    return True

def color_map(countries, colors):
    def backtrack(assignment):
        if len(assignment) == len(countries):
            return assignment
        country = [c for c in countries if c not in assignment][0]
        for color in colors:
            if is_safe(country, color, assignment):
                assignment[country] = color
                result = backtrack(assignment)
                if result:
                    return result
                assignment.pop(country)
        return None
    return backtrack({})

print("Country Color Assignment:", color_map(countries, colors))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = [[1, 2], [2, 3], [3, 3], [4, 5], [5, 6]]
y = [0, 0, 1, 1, 1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = GaussianNB()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print("Predictions:", predictions)
print("Accuracy:", accuracy_score(y_test, predictions))
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
import numpy as np

# Linear Regression
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 2, 3, 4, 5])
linear_model = LinearRegression()
linear_model.fit(X, y)
print("Linear Regression Prediction for input 6:", linear_model.predict([[6]]))

# Logistic Regression
X = np.array([[0], [1], [2], [3], [4]])
y = np.array([0, 0, 0, 1, 1])
logistic_model = LogisticRegression()
logistic_model.fit(X, y)
print("Logistic Regression Prediction for input 5:", logistic_model.predict([[5]]))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = [[0, 0], [1, 1], [2, 2], [3, 3]]
y = [0, 0, 1, 1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print("Predictions:", predictions)
print("Accuracy:", accuracy_score(y_test, predictions))
```
